### 1. Implement the famous FizzBuzz algorithm!


Write a function (or just code without a function around it) that will print into the console integer numbers from 1 to 100.

In addition an algorithm should print in the same row as number specific word (or words) i.e.:

- For each number divisible by three it should print “number-that-is-divisible-by-three” and “Fizz” word (e.g: “3 – Fizz”)
- For each number divisible by five it should print “number-that-is-divisible-by-five” and “Buzz” word (e.g.: “5 – Buzz”)
- If a number is divisible by both three and five it should print “FizzBuzz” word (e.g.: “15 – FizzBuzz”)

```python
def integer_numbers():
    for i in range(1,101):
        # number is divisible by both three and five
        if i % 3 == 0 and i % 5 == 0:
            print(f"{i} - FizzBuzz")
        # number divisible by three
        elif i % 3 == 0:
            print(f"{i} - Fizz")
        # number divisible by five
        elif i % 5 == 0:
            print(f"{i} - Buzz")
        else:
            print(i)
            
integer_numbers()       
        
```

### 2. Implement a system check for new users registering on a GOOGLE-webmail that recently introduced.

Please implement rules as stated below:

Imagine that you're writing application that requires user input and your responsibility is a functionality that will accept and validate user name and email.
Please write a code that will prompt user for both name and email and verify is these fields meet following expectations:


1. User name can’t include any numeric characters (digits)
2. User name can’t exceed 15 characters
3. Email address must start with user name
4. All email letters must be in lower case (here I’d provide hint that, please note that username and email address case might be different)
5. Email address must end with `@google.com`
6. Email address must include underscore (`_`) character
7. Please validate that there are no whitespaces in email address.

In case any of these requirements is not met, print a message that will allow user to easily fix detected issues.

Keep asking for input as long as it's not valid.

Keep in mind that it's quite common for users to accidentally include space at the beginning or end of the input they provide and it would be perfect if code could handle such case, without prompting user to enter input once again.

 

Hints:

- all checks can be implemented using built-in string methods

- you can use one of the built-in loops to keep prompting for input as long as it's not valid

 

Example (happy path):

 

Please provide your name: Foo

Please provide your email address: foo_bar@google.com

 

Example (failure path):

 

Please provide your name: Foo

Please provide your email address:  papayajuice@foo.bar


Whoops! It looks like your email doesn’t meet following requirements:|

```python
def system_check():
    while True:
        # remove whitespaces
        # converting email to lower cases
        
        user_name = input("Please enter your user name: ").strip()
        email = input("Please enter your email address: ").strip().lower()  

        # list with errors
        error_messages = []
        
        ####### USER NAME VALIDATION ###########
        
        # user name - no digits
        if any(number.isdigit() for number in user_name):
            error_messages.append("User name can't include any numeric characters.")

        # user name - length
        if len(user_name) > 15:
            error_messages.append("User name can't exceed 15 characters.")

         ####### EMAIL VALIDATION #############   
            
        # email - starts with user name
        if not email.startswith(user_name.lower()):
            error_messages.append("Email address must start with user name.")

        # email - end with @google.com
        if not email.endswith("@google.com"):
            error_messages.append("Email address must end with '@google.com'.")

        # email - underscore
        if "_" not in email:
            error_messages.append("Email address must include an underscore ('_').")

        # email - no whitespaces
        if " " in email:
            error_messages.append("Email address must not contain any whitespaces.")

        # if error
        if error_messages:
            print("Whoops! It looks like your input doesn’t meet the following requirements:")
            print("\n".join(error_messages))
            continue

        # if no error
        print(f"Your user name and email address have been validated successfully: {user_name} and {email}.")
        break

system_check()
```

### 3. Implement Greatest Common Divisor 

```python
input_a = int(input("Provide first number: "))
input_b = int(input("Provide second number: "))

def find_greatest(input_a, input_b):
    while input_a != input_b:
        if input_a > input_b:
            input_a = input_a - input_b
        else:
            input_b = input_b - input_a
    return input_a
find_greatest(input_a, input_b)
```
