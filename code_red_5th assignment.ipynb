##### 1.
Write a python code to verify if three side (line) lengths provided by user can make a triangle:
Provide three side (line) lengths:

a: 1

b: 2

c: 3

Answer: Can't form a triangle.

 

```python
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))

def triangle(a, b, c):
    if a + b > c and a + c > b and b + c > a:
        print("Can form a triangle.")
    else:
        print("Can't form a triangle.")
        
triangle(a, b, c)
```

##### 2.
Create a robust and user-friendly system for validating pin-codes entered by users.
Develop a Pin-Code Validator code to ensure pin security and adherence to specific criteria:

¬∑       Pin-codes consist only of numbers and are either 4 or 6 digits in length.

¬∑       Enforce a rule preventing consecutive numbers to repeat in the pin-code sequence.

¬∑       Ask for confirmation to check if two entered pin-codes match, indicating successful validation.

 

Good Pin-Codes:

"1234" - 4 digits, no consecutive numbers repeat.

"987654" - 6 digits, no consecutive numbers repeat.

 

Bad Pin-Codes:

"0000" - Consecutive numbers repeated.

"1145" - Consecutive numbers repeated.

"12345" - 5 digits

"34" - 2 digits

"ab12" - 4 digits, contains letters.

```python
pin  = str(input("Provide PIN: "))
def valid_pin(pin):
    
    if not (len(pin) == 4 or len(pin) == 6):
        return False, "Pin-code must be either 4 or 6 digits long."
    
    if not pin.isdigit():
        return False, "Pin-code must consist only of numbers."
    
    for i in range(len(pin) - 1):
        if pin[i] == pin[i + 1]:
            return False, "Pin-code must not contain consecutive repeating numbers."
    
    return True, "Pin-code is valid."

valid_pin(pin)
```

##### 3.
 Develop an interactive "Paper, Scissors, Rock" game for player vs computer gameplay (computer uses choices üòâ).
Count points for human and computer. Stop game after one of players has 3 rounds won.

```python
import random

count_game = 0
count_wins = 3

while count_game < count_wins:
    human_input = input("Enter a choice (rock, paper, scissors): ")

    possible_actions = ["rock", "paper", "scissors"]
    computer_input = random.choice(possible_actions)

    print(f"You chose {human_input}, computer chose {computer_input}.")

    if human_input == computer_input:
        print(f"Both players selected {human_input}. It's a tie!")
    elif human_input == "rock":
        if computer_input == "scissors":
            print("Rock smashes scissors! You win!")
            count_game += 1
        else:
            print("Paper covers rock! You lose.")
            count_game += 1
    elif human_input == "paper":
        if computer_input == "rock":
            print("Paper covers rock! You win!")
            count_game += 1
        else:
            print("Scissors cuts paper! You lose.")
            count_game += 1
    elif human_input == "scissors":
        if computer_input == "paper":
            print("Scissors cuts paper! You win!")
            count_game += 1
        else:
            print("Rock smashes scissors! You lose.")
            count_game += 1

if count_game >= count_wins:
    print("Congratulations! You won the game.")
```

##### 4.
 Create a Python application that displays a random tip from a collection of tips stored in a text file.
 

Create text file:

tips.txt

Tip: Don't put all your eggs in one basket, unless it's an Easter egg hunt.

Tip: Decorating eggs is egg-cellent fun! Just don't end up with egg on your face.

Tip: Don't count your chicks before they hatch. Especially if they're chocolate!

Tip: Easter egg hunts are like life - sometimes you find joy in unexpected places.

```python
tips = [
    "Tip: Don't put all your eggs in one basket, unless it's an Easter egg hunt.",
    "Tip: Decorating eggs is egg-cellent fun! Just don't end up with egg on your face.",
    "Tip: Don't count your chicks before they hatch. Especially if they're chocolate!",
    "Tip: Easter egg hunts are like life - sometimes you find joy in unexpected places."
]

with open('tips.txt', 'w') as file:
    for tip in tips:
        file.write(tip + '\n')

import random

def display_random_tip():
    with open('tips.txt', 'r') as file:
        tips = file.readlines()
        random_tip = random.choice(tips).strip()
        print(random_tip)

display_random_tip()
```

##### 5. Happy Easter
You are the organizer of an Easter egg hunt. There are regular eggs and the chocolate ones hidden all over the place.

When the children finished searching, your assistant counted the finds and included the results in a file egghunt.csv.

You decided that every chocolate egg is worth three points and the regular one, one point.

It appeared that some of found regular eggs laid in bushes since the last Easter, no points for the stinky ones, sorry.

Your job is to analyze and present the results. Pandas will help you with this task.

 

Who is the winner?
How many chocolate eggs was found in total?
Print individual results (kid and his/her score)
Print team results (by family)
Install 'openpyxl' library and use it to print the individual results to 'result.xlsx' file

```python
import pandas as pd


df = pd.read_csv("egghunt.csv")
```

```python
df
```

```python
df['chocolate_points'] = df['chocolate'] * 3
df['regular_points'] = df['regular'] * 1
df['stinky_points'] = df['stinky'] * 0

df['total_points'] = df['chocolate_points'] + df['regular_points'] + df['stinky_points']
winner = df.loc[df['total_points'].idxmax()]
total_chocolate_eggs = df['chocolate'].sum()
team_results = df.groupby('family')['total_points'].sum()

print(f"The winner is {winner['name']} with {winner['total_points']} points!")
print("Total chocolate eggs found:", total_chocolate_eggs)
print("Individual Results:")
print(df[['name', 'family', 'total_points']])

print("Team Results:")
print(team_results)

df[['name', 'family', 'total_points']].to_excel('result.xlsx', index=False)
```

##### 6.
Eggs hunting II: develop a Guessing Game where players find hidden eggs on a 5x5 board.
There should be 10 eggs put randomly on the board.
Allow indefinite number of guesses.
Count number of misses/total attempts.
What is your best result? :)
 

Optional:

Remember if player has already tried specific field.

Draw board after each round (without showing hidden eggs).

Total guesses: 3

Your next guess: B2

 1 2 3 4 5

A . . . . .

B . O . . .

C . . . O .

D . x . x .

E . . . . .

```python
import random

def play_game():
    print("Welcome to the Egg Guessing Game!")
    board = [[' ' for _ in range(5)] for _ in range(5)] 
    player_view = [['.' for _ in range(5)] for _ in range(5)] 
    egg_count = 10
    total_eggs = egg_count  

    
    while egg_count > 0:
        x = random.randint(0, 4)
        y = random.randint(0, 4)
        if board[x][y] != '0':
            board[x][y] = '0'
            egg_count -= 1

    attempts_left = 3
    found_eggs = 0

    while attempts_left > 0:
        print("\n  1 2 3 4 5")
        print(" +---------+")
        row_label = 'A'
        for i in range(5):
            print(f"{row_label}|{'|'.join(player_view[i])}|")
            print(" +---------+")
            row_label = chr(ord(row_label) + 1)

        guess = input("Your next guess (e.g., B2): ").upper()
        guess_x = ord(guess[0]) - ord('A')
        guess_y = int(guess[1]) - 1

        if board[guess_x][guess_y] == '0':  
            print("Congratulations! You found an egg!")
            player_view[guess_x][guess_y] = 'X'
            found_eggs += 1
            if found_eggs == total_eggs:
                print("You've found all the eggs! Congratulations!")
                break
        else:
            if player_view[guess_x][guess_y] == '.':
                player_view[guess_x][guess_y] = 'O'
                attempts_left -= 1
                print(f"Sorry, no egg found. Attempts left: {attempts_left}")
            else:
                print("You've already tried this spot. Try somewhere else.")

        if attempts_left == 0:
            print("Game Over. You've used all your attempts.")
            play_again = input("Do you want to play again? (yes/no): ").lower()
            if play_again == 'yes':
                play_game() 
            break

play_game()
```
